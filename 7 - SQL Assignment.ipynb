```python
#Number 11 Show the 5 most expensive salaries for each team in the year 2014.
query = """SELECT *
    FROM (SELECT salary, teamid, yearid, rank() over (PARITTION BY teamid ORDER BY salary DESC as salaryrank)
    FROM salaries
    WHERE yearid = 2014) ranking
    WHERE salaryrank <= 5
    ORDER BY teamid, salary DESC"""
pd.read_sql_query(query, cnx)
```

```python
#Number 12 Partition the average salaries by team and year, against year. Find players that were paid more than 1 standard deviation above the average salary for that team and year. Show a count by playerid.
query = """SELECT salaries.teamid, salaries.salary, salaries.playerid, salaries.yearid, avgsalary, stddevsalary
FROM salaries
INNER JOIN (SELECT yearid, teamid, AVG(salary) AS "avgsalary", STDDEV(salary) AS "stdsalary"
    FROM salaries
    GROUP BY yearid, teamid) avgsalary ON avgstddev.teamid = salaries.teamid AND avgstddev.yearid = salaries.yearid
    WHERE salaries.salary > avgsalary + stddevsalary
;"""
pd.read_sql_query(query, cnx)
```

```python
#Number 13 Calculate the win percentage. convert w and g into numerics (floats) to do so. (w::numeric, for example)
query = """SELECT teamid, teams.w, teams.l, SUM(w::numeric / (w + l::numeric) * 100)
FROM teams
GROUP BY teamid, teams.w, teams.l;"""
pd.read_sql_query(query, cnx)
```

```python
#Number 14 rank() the total spend by team each year against their actual rank that year. Is there a correlation of spend to performance?
query = """SELECT sums.yearid, sums.teamid, rank() over (PARTITION by sums.yearid ORDER BY sums.add DESC, teamsrank
    FROM (SELECT teamid, yearid, SUM(salary) AS sums
    FROM salaries
    GROUP BY yearid, teamid
    ORDER BY yearid, teamid) as sums
    INNER JOIN teams on teams.yearid = sums.yearid AND
    teams.teamid = sums.teamid
;"""
pd.read_sql_query(query, cnx)
```
