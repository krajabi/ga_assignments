## Background

### Description of Datasets

Dataset 1 (zipcodes): This dataset is one of two datasets that contain observed and expected rates (per 100,000 people) for Agency for Healthcare Research and Quality Prevention Quality Indicators by zipcode. The Agency for Healthcare Research and Quality (AHRQ) Prevention Quality Indicators (PQIs) are a set of population based measures that can be used with hospital inpatient discharge data to identify ambulatory care sensitive conditions. These are conditions where 1) the need for hospitalization is potentially preventable with appropriate outpatient care, or 2) conditions that could be less severe if treated early and appropriately. https://health.data.ny.gov/Health/Hospital-Inpatient-Prevention-Quality-Indicators-P/5q8c-d6xq

Dataset 2 (covered): Health Care Coverage Status (BRFSS): Behavioral Risk Factor Surveillance System (BRFSS) sample data were used to generate annual percentages of non-institutionalized adult (18+) NYS residents with/without health insurance coverage. Health care coverage percentages are provided for 2012 forward. The sample covers between 6,000 and 9,000 completed interviews annually: https://health.data.ny.gov/Health/Health-Care-Coverage-Status-BRFSS-Beginning-2011/g5ce-hdeb.

Dataset 3 (pqi): This is one of two datasets that contain observed and expected rates for Agency for Healthcare Research and Quality Prevention Quality Indicators – Adult (AHRQ PQI) in 2012. This dataset is at the county level.https://health.data.ny.gov/Health/Hospital-Inpatient-Prevention-Quality-Indicators-P/iqp6-vdi4

Dataset 4 (patients): The Statewide Planning and Research Cooperative System (SPARCS) Inpatient De-identified dataset contains discharge level detail on patient characteristics, diagnoses, treatments, services, and charges. https://health.data.ny.gov/Health/Hospital-Inpatient-Discharges-SPARCS-De-Identified/3m9u-ws8e.

Note: Only using simplified dataset from NYC, Manhattan county.
